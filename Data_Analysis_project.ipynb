# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

# Read csv

```python
df=pd.read_csv("F:\Data_analysis_project\zomato.csv\zomato.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df= df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.info()
```

# dropping_duplicates

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

# cleaning rate column

```python
def rate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(rate)
df['rate'].head()
```

# filling null values in rate column

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna('0', inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

# dropping null values

```python
df.dropna(inplace=True)
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plate','listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
df.location.unique()
```

# listed_in(city) and location are there so drop one

```python
df=df.drop(['listed_in(city)'],axis=1)
df.head()
```

```python
df['Cost2plate'].unique()
```

# Removing , from Cost2Plates Column 

```python
def comma(value):
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df['Cost2plate'] = df['Cost2plate'].apply(comma)
df['Cost2plate'].unique()
        
```

# Cleaning Rest Type Column

```python
rest_types = df['rest_type'].value_counts()
rest_types
```

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

# Making Rest Types less than 1000 in frequency as others

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

# cleaning location

```python
df.head()
```

```python
loc = df['location'].value_counts()
loc
```

```python
loclessthan1000 = loc[loc<1000]
loclessthan1000
```

# making less than 1000 location as others

```python
def location(value):
    if value in loclessthan1000:
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(location)
df['location'].value_counts()
        
```
